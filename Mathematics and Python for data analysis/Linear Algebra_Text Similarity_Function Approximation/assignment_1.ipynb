### 2. Произвести токенизацию (разбиение текстов на слова). Удалить пустые слова.

```python
import re
sentences = []
with open('sentences1.txt', 'r') as file_obj:
    words = set([])
    for line in file_obj:
        line = line.lower()
        line = re.split('[^a-z]',line)
        line = ' '.join(line).split()
        sentences.append(line)
for sentence in sentences:
    print(sentence)
```

### 3. Составить список всех слов.


```python
words_dict = {}

i = 0 
for sentence in sentences:
    for word in sentence:
        if word not in words_dict:
            words_dict[word] = i
            i += 1
for word in words_dict:
    print (word, " : ", words_dict[word])
```

```python
framewords = pd.DataFrame(words_dict, range(len(sentences)))
rows, colums = framewords.shape
print(rows, colums)
```

### 6. Найти косинусное расстояние от предложения в самой первой строке до всех остальных.

```python
from scipy.spatial.distance import cosine
```

```python
distance_1_sentence =[]

for i in range(rows):
    distance_1_sentence.append(cosine(framewords.iloc[0],framewords.iloc[i] ))

print( distance_1_sentence)
```

```python
distance_1_sentence_copy = list(distance_1_sentence)
twoClosestValues = [[-1, 0], [-1, 0]]

distance_1_sentence.remove(min(distance_1_sentence_copy))

for i in range(2):
    twoClosestValues[i][1] = min(distance_1_sentence_copy)

    for j in range(len(distance_1_sentence)):
        if twoClosestValues[i][1] == distance_1_sentence[j]:
            
            twoClosestValues[i][0] = j
    distance_1_sentence_copy.remove(min(distance_1_sentence_copy))
    
twoClosestValues = sorted(twoClosestValues)
twoClosestValues
```
