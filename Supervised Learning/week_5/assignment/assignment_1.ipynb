```python
# импортируем необходимые библиотеки
import pandas as pd
import numpy as np
from sklearn import model_selection
from sklearn import datasets
from sklearn.naive_bayes import  BernoulliNB, MultinomialNB, GaussianNB

```

```python
# загрузим данные 

digits_data = pd.DataFrame(datasets.load_digits().data)
digits_target = pd.DataFrame(datasets.load_digits().target)
cancer_data = pd.DataFrame(datasets.load_breast_cancer().data)
cancer_data.columns = datasets.load_breast_cancer().feature_names
cancer_target = pd.DataFrame(datasets.load_breast_cancer().target)

```

```python
cancer_data.head()
```

```python
digits_data.head()
```

```python
bayeses = [BernoulliNB, MultinomialNB, GaussianNB]

for bayes in bayeses:
    clf = bayes()
    clf.fit(digits_data, digits_target)
    digits_score = model_selection.cross_val_score(clf, X = digits_data, y = digits_target).mean()
    cancer_score = model_selection.cross_val_score(clf, X = cancer_data, y = cancer_target).mean()
    print("{} : \n {} - {} ; {} - {}".format(str(bayes),'digits_score',str(digits_score),'cancer_score', str(cancer_score)))
```
